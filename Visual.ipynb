```python
### Importing libraries and necessary resources
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt

df = pd.read_csv("attacks.csv",encoding = "ISO-8859-1") #To load the data csv
surf_usa = pd.read_csv("USA_Surfers.csv",encoding = "ISO-8859-1")
surf_aus = pd.read_csv("Surfing in Australia.csv",encoding = "ISO-8859-1")
champ = pd.read_csv("WSL_F_Winners.csv" ,encoding = "ISO-8859-1")
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("whitegrid")
```

# Visualizaciones finales

# Distribución origen de ataques en USA y AUS

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
sns.histplot(data = surf_usa,x="Year",hue = "Type",multiple="stack",palette="deep",ax=axs[0])
sns.histplot(data = surf_aus,x="Year",hue = "Type",multiple="stack",palette="deep",ax=axs[1])
```

## Podemos observar que tanto en Estados Unidos como en Australia la gran mayoría de los ataques a surfistas son de la categoría "Unprovoked", por lo que podemos extraer de esta información, que los tiburones tienen cierto gusto por la carne de los surfistas independientemente de su procedencia. Al menos no son racistas.

# Distribución ataques totales

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
sns.histplot(data=surf_usa,x="Year",bins=9,kde=True,palette="deep",ax=axs[0])
sns.histplot(data=surf_aus,x="Year",bins=10,kde=True,palette="deep",ax=axs[1])
```

## Con el análisis histórico de los ataques a este grupo de humanos, observamos que los tiburones tienen cierta preferencia por los surfistas americanos a lo largo de la historia. Deliciosos


# Distribución ataques entre USA, AUS y por sexo

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
sns.swarmplot(data = surf_usa,x="Year",y="Sex",hue="Fatal",palette="deep",ax=axs[0])
sns.swarmplot(data = surf_aus,x="Year",y="Sex",hue="Fatal",palette="deep",ax=axs[1])
```

## Con este último gráfico podemos confirmar que la carne de surfista americano es la absoluta ganadora y la más valorada entre los escualos, seguida de la de los surfistas australianos. 

# Comparativa fatalidad ataques USA/AUS

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,10))
sns.countplot(x=champ.Countries,palette="deep",ax=axs[1])
sns.histplot(data = surf_usa,x="Year",hue = "Fatal",multiple="stack",palette="deep",ax=axs[0])
sns.histplot(data = surf_aus,x="Year",hue = "Fatal",multiple="stack",palette="deep",ax=axs[2])
```
